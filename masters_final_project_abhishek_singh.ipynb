```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import plotly.express as px

def load_and_validate_data(filepath):
    """Load data and validate required columns exist."""
    df = pd.read_csv(filepath)
    required_columns = [
        'Happiness_Score', 'Country', 'Year',
        'GDP_per_Capita', 'Social_Support',
        'Healthy_Life_Expectancy', 'Freedom'
    ]
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"Missing columns: {missing_columns}")
    return df

def calculate_correlations(df, indicators, min_samples=10):
    """Calculate Pearson correlations between happiness score and indicators."""
    correlations = {}
    for indicator in indicators:
        valid_data = df[['Happiness_Score', indicator]].dropna()
        if len(valid_data) >= min_samples:
            corr, p_value = pearsonr(valid_data['Happiness_Score'], valid_data[indicator])
            correlations[indicator] = {
                'correlation': corr,
                'p_value': p_value,
                'sample_size': len(valid_data)
            }
        else:
            print(f"Warning: Insufficient data for {indicator} ({len(valid_data)} samples)")
    return correlations

def create_plotly_scatter_plots(df, indicators, correlations, colors):
    """Create interactive Plotly scatter plots with trendlines."""
    for i, indicator in enumerate(indicators):
        if indicator not in correlations:
            continue
        clean_data = df[['Happiness_Score', indicator, 'Country', 'Year']].dropna()
        fig = px.scatter(
            clean_data,
            x=indicator,
            y='Happiness_Score',
            hover_name='Country',
            trendline='ols',
            title=f'Happiness Score vs {indicator.replace("_", " ")} (r = {correlations[indicator]["correlation"]:.3f})',
            labels={indicator: indicator.replace('_', ' '), 'Happiness_Score': 'Happiness Score'}
        )
        # Set all marker colors to the custom palette
        fig.update_traces(marker=dict(size=8, color=colors[i], line=dict(width=1, color='white')))
        fig.show()


def print_color_palette(indicators, colors):
    print("Custom Happiness Color Palette:")
    for indicator, color in zip(indicators, colors):
        print(f"{indicator.replace('_', ' ')}: {color}")
    print("-" * 50)

def print_correlation_summary(correlations):
    print("\nCorrelation Summary:")
    print("-" * 50)
    for indicator, stats in correlations.items():
        significance = "***" if stats['p_value'] < 0.001 else \
                       "**" if stats['p_value'] < 0.01 else \
                       "*" if stats['p_value'] < 0.05 else ""
        print(f"{indicator.replace('_', ' ')}: r = {stats['correlation']:.3f}{significance} (n = {stats['sample_size']})")
    print("\nSignificance: *** p<0.001, ** p<0.01, * p<0.05")

# ---------- Jupyter Notebook Usage Starts Here ----------

# File and indicator setup
DATA_FILE = 'world_happiness_report.csv'
INDICATORS = ['GDP_per_Capita', 'Social_Support', 'Healthy_Life_Expectancy', 'Freedom']
COLORS = [
    "#419DD7", 
    "#FFB14E",  
    "#E43C32",  
    "#8663AC"   
]

# Load, validate, and analyze the data
df = load_and_validate_data(DATA_FILE)
correlations = calculate_correlations(df, INDICATORS)

# Uncomment below to print palette and correlation summary if desired
#print_color_palette(INDICATORS, COLORS)
#print_correlation_summary(correlations)

# Create interactive scatter plots
create_plotly_scatter_plots(df, INDICATORS, correlations, COLORS)
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    try:
        # Some country names may need manual fixes
        country_corrections = {
            'United States': 'United States of America',
            'Russia': 'Russian Federation',
            'South Korea': 'Korea, Republic of',
            'North Korea': 'Korea, Democratic People\'s Republic of',
            'Czechia': 'Czech Republic',
            'Vietnam': 'Viet Nam',
            'Iran': 'Iran, Islamic Republic of',
            'Syria': 'Syrian Arab Republic',
            'Laos': 'Lao People\'s Democratic Republic',
            'Moldova': 'Moldova, Republic of',
            'Tanzania': 'Tanzania, United Republic of',
            'Venezuela': 'Venezuela, Bolivarian Republic of',
            'Bolivia': 'Bolivia, Plurinational State of',
            'Brunei': 'Brunei Darussalam',
            'Palestine': 'Palestine, State of',
            'Congo (Brazzaville)': 'Congo',
            'Congo (Kinshasa)': 'Congo, The Democratic Republic of the',
            'Ivory Coast': "Côte d'Ivoire"
        }
        country_name = country_corrections.get(country_name, country_name)
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except Exception:
        return None

df['Continent'] = df['Country'].apply(country_to_continent)

fig = px.violin(
    df,
    x='Continent',
    y='Happiness_Score',
    color='Continent',
    color_discrete_sequence=[
        "#E4572E", "#4E79A7", "#76B7B2", "#59A14F", "#F28E2B", "#B07AA1"
    ],
    box=True,  # Show box inside violin
    points="all",
    title="Happiness Score Distribution by Continent"
)
fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Load the data
df = pd.read_csv('world_happiness_report.csv')

# Select key happiness factors for clustering
happiness_factors = ['GDP_per_Capita', 'Social_Support', 'Healthy_Life_Expectancy', 
                    'Freedom', 'Generosity', 'Corruption_Perception']

# Create a clean dataset for clustering
cluster_data = df[['Country', 'Year', 'Happiness_Score'] + happiness_factors].dropna()

print(f"Dataset shape for clustering: {cluster_data.shape}")
print(f"Countries available: {cluster_data['Country'].nunique()}")
print(f"Years covered: {cluster_data['Year'].min()} - {cluster_data['Year'].max()}")

# Let's use the most recent year for each country to avoid duplicates
latest_data = cluster_data.groupby('Country').last().reset_index()
print(f"\nLatest data shape: {latest_data.shape}")

# Prepare features for clustering
X = latest_data[happiness_factors]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"\nFeatures used for clustering:")
for i, feature in enumerate(happiness_factors):
    print(f"  {i+1}. {feature}")
```

```python
# Find optimal number of clusters using elbow method and silhouette score
k_range = range(2, 8)
inertias = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot elbow curve and silhouette scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Elbow method
ax1.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia (Within-cluster sum of squares)')
ax1.set_title('Elbow Method for Optimal k')
ax1.grid(True, alpha=0.3)

# Silhouette scores
ax2.plot(k_range, silhouette_scores, 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score for Different k')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print scores
print("Cluster Evaluation:")
print("k\tInertia\t\tSilhouette Score")
print("-" * 35)
for k, inertia, sil_score in zip(k_range, inertias, silhouette_scores):
    print(f"{k}\t{inertia:.2f}\t\t{sil_score:.3f}")

# Choose optimal k (highest silhouette score)
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"\nOptimal number of clusters: {optimal_k} (highest silhouette score: {max(silhouette_scores):.3f})")
```

```python
# Apply K-Means with optimal number of clusters
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to the dataset
latest_data['Cluster'] = cluster_labels
latest_data['Cluster_Label'] = ['Cluster ' + str(i) for i in cluster_labels]

# Apply PCA for 2D visualization
pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_scaled)

# Apply PCA for 3D visualization
pca_3d = PCA(n_components=3, random_state=42)
X_pca_3d = pca_3d.fit_transform(X_scaled)

# Add PCA coordinates to dataset
latest_data['PCA1'] = X_pca_2d[:, 0]
latest_data['PCA2'] = X_pca_2d[:, 1]
latest_data['PCA1_3D'] = X_pca_3d[:, 0]
latest_data['PCA2_3D'] = X_pca_3d[:, 1]
latest_data['PCA3_3D'] = X_pca_3d[:, 2]

print("PCA Explained Variance Ratio:")
print(f"2D PCA: PC1={pca_2d.explained_variance_ratio_[0]:.3f}, PC2={pca_2d.explained_variance_ratio_[1]:.3f}")
print(f"Total variance explained (2D): {sum(pca_2d.explained_variance_ratio_):.3f}")
print(f"3D PCA: PC1={pca_3d.explained_variance_ratio_[0]:.3f}, PC2={pca_3d.explained_variance_ratio_[1]:.3f}, PC3={pca_3d.explained_variance_ratio_[2]:.3f}")
print(f"Total variance explained (3D): {sum(pca_3d.explained_variance_ratio_):.3f}")

# Show cluster assignments
print(f"\nCluster Assignments:")
for cluster in range(optimal_k):
    countries = latest_data[latest_data['Cluster'] == cluster]['Country'].tolist()
    print(f"Cluster {cluster}: {countries}")
```

```python
# 1. PLOTLY: Interactive 2D scatter plot with clusters
fig_2d = px.scatter(latest_data, 
                    x='PCA1', y='PCA2',
                    color='Cluster_Label',
                    hover_name='Country',
                    hover_data=['Happiness_Score', 'GDP_per_Capita', 'Social_Support'],
                    title='Country Clusters Based on Happiness Factors (2D PCA)',
                    labels={'PCA1': f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)',
                           'PCA2': f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)'},
                    color_discrete_sequence=['#E74C3C', '#3498DB', '#2ECC71'])

# Add cluster centers
centers_2d = pca_2d.transform(kmeans.cluster_centers_)
for i, center in enumerate(centers_2d):
    fig_2d.add_trace(go.Scatter(x=[center[0]], y=[center[1]], 
                               mode='markers+text',
                               marker=dict(size=15, color='black', symbol='x'),
                               text=[f'Center {i}'],
                               textposition='top center',
                               name=f'Cluster {i} Center',
                               showlegend=False))

fig_2d.update_layout(width=800, height=600)
fig_2d.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use a clean, modern Seaborn style
sns.set_theme(context='notebook', style='whitegrid', font_scale=1.2)

# Updated cute pastel color palette (from your network diagram)
pastel_colors = ['#a8dadc', '#ffe066', '#f7a072', '#70c1b3', '#bdb2ff', '#fdffb6', '#ffc6ff']
pca_bar_colors = ['#FFD700', '#33C6DD']

fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Plot 1: 2D PCA scatter plot
for i, cluster in enumerate(range(optimal_k)):
    cluster_data = latest_data[latest_data['Cluster'] == cluster]
    axes[0,0].scatter(
        cluster_data['PCA1'], cluster_data['PCA2'], 
        c=cluster_colors[i % len(cluster_colors)], label=f'Cluster {cluster}', 
        alpha=0.9, s=120, edgecolor='white', linewidth=1.5
    )
    # Add country labels
    for _, row in cluster_data.iterrows():
        axes[0,0].annotate(
            row['Country'], (row['PCA1'], row['PCA2']),
            xytext=(6, 5), textcoords='offset points', fontsize=9, fontweight='bold', color='#444'
        )

# Add cluster centers
centers_2d = pca_2d.transform(kmeans.cluster_centers_)
axes[0,0].scatter(
    centers_2d[:, 0], centers_2d[:, 1], 
    c='#000', marker='*', s=400, linewidths=2, label='Centers', zorder=5
)
axes[0,0].set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} variance)', fontsize=13)
axes[0,0].set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} variance)', fontsize=13)
axes[0,0].set_title('Country Clusters (2D PCA)', fontsize=15, fontweight='bold')
axes[0,0].legend(frameon=True, fontsize=11)
axes[0,0].grid(True, alpha=0.25, linestyle='--')
axes[0,0].set_axisbelow(True)
axes[0,0].set_facecolor("#FAFAFA")
sns.despine(ax=axes[0,0])

# Plot 2: Cluster characteristics heatmap
cluster_means = latest_data.groupby('Cluster')[happiness_factors].mean()
sns.heatmap(
    cluster_means.T, annot=True, fmt=".2f", cmap='YlGnBu', center=0,
    ax=axes[0,1], cbar_kws={'label': 'Standardized Value'},
    linewidths=0.6, linecolor='white', square=True, annot_kws={"size": 12, "weight": "bold"}
)
axes[0,1].set_title('Cluster Characteristics Heatmap', fontsize=15, fontweight='bold')
axes[0,1].set_xlabel('Cluster', fontsize=12)
axes[0,1].set_ylabel('')

# Plot 3: Happiness score by cluster (box + swarm)
sns.boxplot(
    data=latest_data, x='Cluster', y='Happiness_Score', 
    palette=cluster_colors, ax=axes[1,0], fliersize=0, width=0.55, linewidth=1.3
)
sns.swarmplot(
    data=latest_data, x='Cluster', y='Happiness_Score', 
    color='#111', alpha=0.67, ax=axes[1,0], size=6
)
axes[1,0].set_title('Happiness Score Distribution by Cluster', fontsize=15, fontweight='bold')
axes[1,0].set_xlabel('Cluster', fontsize=12)
axes[1,0].set_ylabel('Happiness Score', fontsize=12)
axes[1,0].set_facecolor("#F7FAFD")
axes[1,0].grid(axis='y', alpha=0.22)
sns.despine(ax=axes[1,0])

# Plot 4: Feature importance (PCA loadings)
feature_importance = pd.DataFrame(
    pca_2d.components_.T,
    columns=['PC1', 'PC2'],
    index=happiness_factors
)
feature_importance.plot(
    kind='bar', ax=axes[1,1], color=pca_bar_colors, edgecolor='white', width=0.77
)
axes[1,1].set_title('Feature Importance in PCA Components', fontsize=15, fontweight='bold')
axes[1,1].set_xlabel('Features', fontsize=12)
axes[1,1].set_ylabel('PCA Loading', fontsize=12)
axes[1,1].tick_params(axis='x', rotation=40)
axes[1,1].legend(frameon=True)
axes[1,1].set_axisbelow(True)
axes[1,1].grid(True, alpha=0.2, linestyle='--')
axes[1,1].set_facecolor("#FBFBFB")
sns.despine(ax=axes[1,1])

plt.tight_layout(pad=2.5)
plt.suptitle("Cluster Analysis: Cute, Clear & Insightful! 🌈", fontsize=20, fontweight='heavy', color='#444', y=1.04)
plt.show()
```

```python
# 4. Detailed cluster analysis and interpretation
print("🎯 CLUSTER ANALYSIS RESULTS")
print("=" * 50)

# Cluster statistics
for cluster in range(optimal_k):
    cluster_countries = latest_data[latest_data['Cluster'] == cluster]
    print(f"\n📊 CLUSTER {cluster}:")
    print(f"Countries: {', '.join(cluster_countries['Country'].tolist())}")
    print(f"Average Happiness Score: {cluster_countries['Happiness_Score'].mean():.2f}")
    
    print("\nKey Characteristics:")
    for factor in happiness_factors:
        avg_value = cluster_countries[factor].mean()
        overall_avg = latest_data[factor].mean()
        if avg_value > overall_avg * 1.1:
            status = "HIGH ⬆️"
        elif avg_value < overall_avg * 0.9:
            status = "LOW ⬇️"
        else:
            status = "AVERAGE ➡️"
        print(f"  • {factor}: {avg_value:.2f} ({status})")

# Feature correlations with happiness
print(f"\n🔗 FEATURE CORRELATIONS WITH HAPPINESS:")
print("-" * 40)
correlations = latest_data[happiness_factors + ['Happiness_Score']].corr()['Happiness_Score'].drop('Happiness_Score')
for factor, corr in correlations.sort_values(ascending=False).items():
    print(f"{factor}: {corr:+.3f}")

# PCA interpretation
print(f"\n🔍 PCA COMPONENT INTERPRETATION:")
print("-" * 35)
feature_loadings = pd.DataFrame(
    pca_2d.components_.T,
    columns=['PC1', 'PC2'],
    index=happiness_factors
)

print("PC1 (Economic & Social Development):")
pc1_sorted = feature_loadings['PC1'].abs().sort_values(ascending=False)
for factor in pc1_sorted.index[:3]:
    loading = feature_loadings.loc[factor, 'PC1']
    print(f"  • {factor}: {loading:+.3f}")

print("\nPC2 (Governance & Freedom):")
pc2_sorted = feature_loadings['PC2'].abs().sort_values(ascending=False)
for factor in pc2_sorted.index[:3]:
    loading = feature_loadings.loc[factor, 'PC2']
    print(f"  • {factor}: {loading:+.3f}")
```
